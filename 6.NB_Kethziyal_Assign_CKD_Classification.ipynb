```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
indep=dataset[[ 'age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
indep
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
#GUASSIAN NB:
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
param_grid = {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]}
grid = GridSearchCV(GaussianNB(), param_grid, cv=5)
grid.fit(X_train, y_train)


```

```python

re=grid.cv_results_

grid_predictions=grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)


from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)

print(" GaussianNB :\n",cm)
print("GaussianNB :\n",clf_report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import ComplementNB
classifier=ComplementNB()
param_grid = {'alpha': [0.1, 0.5, 1.0, 2.0],'force_alpha':[True, False],
             'fit_prior': [True, False],'norm': [True,False] }

grid = GridSearchCV(ComplementNB(), param_grid, cv=5,scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import KBinsDiscretizer

kb = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')

X_train = kb.fit_transform(X_train)
X_test = kb.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import CategoricalNB
classifier=CategoricalNB()
param_grid = {'alpha': [0.1, 0.5, 1.0, 2.0],'fit_prior': [True, False],'force_alpha':[True, False]}
grid = GridSearchCV(CategoricalNB(), param_grid, cv=5)
grid.fit(X_train, y_train)




```

```python
re=grid.cv_results_

grid_predictions=grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)


from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)

print("CategoricalNB :\n",cm)
print("CategoricalNB :\n",clf_report)
```
